## For the moment, let's assume that .txt are pretraining data and .json are SFT data.

```python
pt_fnames = [fname for fname in os.listdir() if osp.splitext(fname)[1] == ".txt"]
it_fnames = [fname for fname in os.listdir() if osp.splitext(fname)[1] == ".json"]
```

```python
set([osp.splitext(fname)[1] for fname in os.listdir()])
```

```python
cur_fnames = [fname for fname in os.listdir() if osp.isfile(fname)]
```

```python
from pprint import pprint

for fname in cur_fnames:
    ext = osp.splitext(fname)[1]
    with open(fname, "r") as f:
        if ext == ".txt":
            examples = [line for line in f]
        elif ext == ".json":
            data = json.load(f)
            if isinstance(data, list):
                examples = data
            elif isinstance(data, dict):
                examples = list(data.items())
            else:
                raise NotImplementedError(f"Invalid data type: {type(data)}")
        else:
            raise NotImplementedError(f"Invalid extension: {ext}")

        print(fname.center(80, "-"))
        pprint(examples[0])
        print("-" * 80, "\n")
```
